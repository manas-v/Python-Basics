## 8.	Write python program using Pandas

```python
import pandas as pd
```

```python
df = pd.read_csv("ab - Sheet1.csv")
print(df.shape)
df.head()
```

#### a.	To count the NaN values in a dataset.

```python
df.isnull().sum()
```

```python
df.notnull().sum()
```

#### b. To handle missing values in a dataset.

```python
#Drop All Rows with any Null/NaN/NaT Values

a = df.dropna()
print(a.shape)
a
```

```python
#Drop All Columns with Any Missing Value

b = df.dropna(axis=1)
print(b.shape)
b.head()
```

```python
#Drop Row/Column Only if All the Values are Null

c = df[df.isna().any(axis=1)]
c
```

```python
c1 = c.dropna(how='all', axis=1)
print(c1.shape)
c1.head()
```

```python
#DataFrame Drop NA based on Rows/Columns

d = df.dropna(subset=['Age'])
print(d.shape)
d.head()
```

```python
#Find the mean,median and mode values in the Age column
mean=df['Age'].mean()
median=df['Age'].median()
mode=df['Age'].mode()
```

```python
# Replace Null Values (np.nan) with mean
df['Age'] = df['Age'].replace(np.nan, mean) #can be done for mean, median, mode
df.head()
```

```python
# Bakward fill
e = df.copy()
e['Cabin'] = e['Cabin'].fillna(method='bfill')
e.head()
```

```python
# Forward fill
f = df.copy()
f['Cabin'] = f['Cabin'].fillna(method='ffill')
f.head()
```

```python
# Most frequent values
df['Cabin'].value_counts()
```

```python
g = df.copy()
g['Cabin'] = g['Cabin'].fillna(g['Cabin'].value_counts().index[0])
g.head()
```

#### c.	To remove duplicate values in a dataset.

```python
data = {'col1': [1, 1, 3, 3], 'col2': [1, 1, 3, 3], 'col3': [2, 2, 6, 5], 'col4': [2, 2, 6, 6]}

df = pd.DataFrame(data)
print('Source DataFrame:\n', df)

df.duplicated()
```

```python
#Drop Duplicate Rows Keeping the First One

df1 = df.copy()
dropped_df1 = df1.drop_duplicates()
print(dropped_df1)
```

```python
#Drop Duplicates and Keep Last Row

df2 = df.copy()
dropped_df2 = df2.drop_duplicates(keep='last')
print(dropped_df2)
```

```python
#Delete All Duplicate Rows from DataFrame

df3 = df.copy()
dropped_df3 = df3.drop_duplicates(keep=False)
print(dropped_df3)
```

```python
#Identify Duplicate Rows based on Specific Columns

df4 = df.copy()
dropped_df4 = df4.drop_duplicates(subset=['col1', 'col2'])
print(dropped_df4)
```

```python
#Remove Duplicate Rows in place

df5 = df.copy()
dropped_df5 = df5.drop_duplicates(inplace=True)
print(dropped_df5)
```
